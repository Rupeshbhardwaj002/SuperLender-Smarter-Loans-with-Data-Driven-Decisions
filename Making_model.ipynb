Here we imported pandas as pd libraray. And after this we downloaded our required datasets which we uploaded on github.


```python
import pandas as pd
train_previous = pd.read_csv('https://raw.githubusercontent.com/Rupeshbhardwaj002/SuperLender-Smarter-Loans-with-Data-Driven-Decisions/refs/heads/main/trainprevloans.csv')
train_perf = pd.read_csv('https://raw.githubusercontent.com/Rupeshbhardwaj002/SuperLender-Smarter-Loans-with-Data-Driven-Decisions/refs/heads/main/trainperf.csv')
train_demo = pd.read_csv('https://raw.githubusercontent.com/Rupeshbhardwaj002/SuperLender-Smarter-Loans-with-Data-Driven-Decisions/refs/heads/main/traindemographics.csv')
test_previous = pd.read_csv('https://raw.githubusercontent.com/Rupeshbhardwaj002/SuperLender-Smarter-Loans-with-Data-Driven-Decisions/refs/heads/main/testprevloans.csv')
test_perf = pd.read_csv('https://raw.githubusercontent.com/Rupeshbhardwaj002/SuperLender-Smarter-Loans-with-Data-Driven-Decisions/refs/heads/main/testperf.csv')
test_demo = pd.read_csv('https://raw.githubusercontent.com/Rupeshbhardwaj002/SuperLender-Smarter-Loans-with-Data-Driven-Decisions/refs/heads/main/testdemographics.csv')
```

Now we mered our dataset in one file from the other 3 training files that includes train previous, train demographics and train perf they all contains a unique key customer id which we used to connect them.

```python
temp_data = pd.merge(train_previous, train_perf, on='customerid',how='inner')
df = pd.merge(temp_data, train_demo, on='customerid', how='inner')
# print(df.shape)
df.head()
```

Here we merged them all in one file df.

```python
df.info()
```

We have drroped column  df['bank_branch_clients'] because a lot of null values in this column.

```python
df = df.drop(['bank_branch_clients'], axis=1)
print(df.columns)
```

```python
df['approveddate_x'] = pd.to_datetime(df['approveddate_x'], errors='coerce')
df['birthdate'] = pd.to_datetime(df['birthdate'], errors='coerce')
df['creationdate_x'] = pd.to_datetime(df['creationdate_x'], errors='coerce')
df['closeddate'] = pd.to_datetime(df['closeddate'], errors='coerce')

# Age at time of loan
df['age'] = (df['approveddate_x'] - df['birthdate']).dt.days // 365

# Loan duration
df['loan_duration_days'] = (df['closeddate'] - df['approveddate_x']).dt.days

# Time since account creation
df['days_since_creation'] = (pd.Timestamp.now() - df['creationdate_x']).dt.days

# Month and weekday of approval
df['approval_month'] = df['approveddate_x'].dt.month
df['approval_weekday'] = df['approveddate_x'].dt.weekday

df.drop(['approveddate_x', 'birthdate', 'creationdate_x', 'closeddate'], axis=1, inplace=True)
```

USING LOOP FOR ENCODING BECAUSE A LOT OF WORK INCREASE IF WE MANUALLY DO EVRY FEATURE SPECIALLY IN MORE LARGE DATASETS.

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for col in df.columns:
  if(df[col].dtype == object):
    df[col] = encoder.fit_transform(df[col])
```

NOW WE WILL DIVIDE OUR DATASET IN X AND Y . AS WE KNOW TEST DATA ONLY CONATAINS INPUTS HENCE IT DOESN'T NEEDS TO BE ANYTHING CHANGED. and we will also drop the columns here that has low coorelation that we find out before.

```python
X = df[['loannumber_x', 'loanamount_x',
       'totaldue_x', 'termdays_x', 'referredby_x', 'firstduedate',
       'firstrepaiddate', 'systemloanid_y', 'loannumber_y', 'approveddate_y',
       'creationdate_y', 'loanamount_y', 'totaldue_y', 'termdays_y',
       'referredby_y', 'bank_account_type', 'longitude_gps', 'bank_name_clients',
        'employment_status_clients','level_of_education_clients', 'age',
        'loan_duration_days','days_since_creation']]
Y = df['good_bad_flag']
```

```python
X.head()
```

we are doing standard scaling because in some of the feature there is a huge diffrence.

```python
from sklearn.preprocessing import StandardScaler
d = StandardScaler()
X = d.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

First we will try with the logistic regression because it is simple and fast for classification tasks

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression(max_iter=2500)
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import recall_score, f1_score, roc_auc_score, accuracy_score, precision_score

print("Accuracy:", accuracy_score(Y_test, y_pred))
print("Precison:", precision_score(Y_test, y_pred))
print("Recall:", recall_score(Y_test, y_pred))
print("F1 Score:", f1_score(Y_test, y_pred))
print("ROC AUC:", roc_auc_score(Y_test, model.predict_proba(X_test)[:, 1]))
```

The metrices result is not like our expectations specially ROC and AUC score which is near to 0.5 which means our model is weak in making diffrence between bad and good loans

```python
from sklearn.metrics import roc_curve, precision_recall_curve
import matplotlib.pyplot as plt

y_scores = model.predict_proba(X_test)[:, 1]

fpr, tpr, _ = roc_curve(Y_test, y_scores)
plt.plot(fpr, tpr)
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

Now we will use boosting classifier - random forest its is a ensemble method technique. it is a strong learner made from many weak leanere(Decison tree) it minimizes the variance and bias but its little slow because of training in sequentail we are also using gridsearch cv to find best result giving parameters of our random forest.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
}
grid_search = GridSearchCV(estimator=rf,
                           param_grid=param_grid,
                           cv=5,
                           scoring='roc_auc',
                           n_jobs=-1,
                           verbose=2)


grid_search.fit(X_train, Y_train)
print("Best:", grid_search.best_params_)
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)
```

Here our metirces are metting our exxpectations. Which means our model is working fine hence we will finalize random forest if we want we can use stacking also because it is more powerful than boosting or other classification techniques.

```python
print("Accuracy:", accuracy_score(Y_test, y_pred))
print("Precison:", precision_score(Y_test, y_pred))
print("Recall:", recall_score(Y_test, y_pred))
print("F1 Score:", f1_score(Y_test, y_pred))
print("ROC AUC:", roc_auc_score(Y_test, best_rf.predict_proba(X_test)[:, 1]))
```
